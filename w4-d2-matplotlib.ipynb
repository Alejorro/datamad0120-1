```python
def square(L=1, c=(0,0)):
    square_points = np.asarray([
        [0,0],
        [L,0],
        [L,L],
        [0,L],
        [0,0]
    ])
    return square_points + np.asarray(c)

s = square(5,(10,10))   
plt.plot(s[:,0],s[:,1],'o-')
```

```python
def square(L=1, c=(0,0)):
    square_points = np.asarray([
        [0,0],
        [L,0],
        [L,L],
        [0,L],
        [0,0]
    ])
    return (square_points + (np.asarray(c)-L/2)).T

s = square(5,(0,0))
print(s)
plt.plot(*s,'o-')
```

```python
#https://es.wikipedia.org/wiki/Matriz_de_rotaci%C3%B3n
import math
def square(L=1, c=(0,0),angle=0):
    angle = np.pi*angle/180 
    R = np.asarray([
        [np.cos(angle),-np.sin(angle)],
        [np.sin(angle),np.cos(angle)]
    ])
#    print(R)
    square_points = np.asarray([
        [0,0],
        [L,0],
        [L,L],
        [0,L],
        [0,0]
    ])
    return (np.dot(square_points+(np.asarray(c)-L/2),R)).T

plt.figure(figsize=(5,5))
plt.plot(*square(5,(0,0),angle=45),'o-')
```

```python
with plt.xkcd():
    plt.figure(figsize=(5,5))
    for L in range(1,10):
        plt.plot(*square(L,(0,0),angle=360/L))

    # https://stackoverflow.com/questions/4700614/how-to-put-the-legend-out-of-the-plot
    plt.legend(range(10),bbox_to_anchor=(1.0, 1.0))
```
