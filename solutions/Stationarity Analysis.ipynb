```python
import numpy as np
from statsmodels.tsa.stattools import adfuller
stationaryCheck = lambda X: "Not-Stationary" if adfuller(X)[1] > 0.05 else "Stationary"
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x1 = np.linspace(0,1,100)
y1 = np.sin(2*np.pi*x1)

plt.plot(x1,y1)
plt.title(stationaryCheck(x1))
```

```python
x2 = np.linspace(0,1,100)
y2 = np.sin(2*np.pi*2000*x2)

plt.plot(x2,y2,'o')
plt.title(stationaryCheck(x2))
```

```python
import pandas as pd
pd.plotting.lag_plot(pd.Series(x1))
```

```python
## NOTE: This is stationary because we have a RANDOM noise added to the signal
y2 = np.sin(2*np.pi*2000*x2)
w = y2 + 2*np.random.random(len(x2))
plt.plot(x2,w)
plt.title(stationaryCheck(w))
```

```python
import pandas as pd
pd.plotting.lag_plot(pd.Series(w))
```

```python
from scipy.fftpack import fft
mod = np.abs(fft(w))
mod = mod[0:len(mod)//2]
plt.subplot(1,2,2)
plt.plot(mod)
```
