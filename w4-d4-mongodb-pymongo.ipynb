```python
# Queries
# 0 - {cuisine:{"$eq":"American"}, name:/Rivi/}
# 1 - {name:/Rivi/,"$or":[{cuisine:"Bakery"},{cuisine:"American"}]}
# 2 - {name:/Rivi/,cuisine:{"$in":["American","Bakery"]} }
# 3 - {cuisine:{"$ne":"American"}, name:/Rivi/}
# 4 - {restaurant_id: { $exists: false }}
# 5 - {grades:{$elemMatch:{score:8}}}
# 6 - {grades:{$elemMatch: {score:{"$gte":50}} }}

# 7 - Restaurantes con todas las notas como m√≠nimo 50, 
#     que tengan minimo 1 nota, y que el campo de la nota no sea null
'''
{
  "$and": [
    {
      "grades": {
        "$not": {
          "$elemMatch": {
            "score": {
              "$lte": 50
            }
          }
        }
      }
    },
    {
      "grades.score": {
        "$not": {
          "$type": "null"
        }
      }
    },
    {
      "grades": {
        "$not": {
          "$size": 0
        }
      }
    }
  ]
}
'''
```

```python
class Vector:
    def __init__(self,*args):
        self.coord = args
    def __repr__(self):
        return f"Vector({','.join(map(lambda x: str(x),self.coord))})"
    def __add__(self,b):
        #return Vector(*[b.coord[i]+c for i,c in enumerate(self.coord)])
        return Vector(*map(lambda s:s[0]+s[1],zip(self.coord,b.coord)))


a = Vector(1,2,3)
b = Vector(4,5,6)
print(a,b)
print(a+b)



    
```

```python
from pandas.io.json import json_normalize

cursor = db["ny"].find({"$or":[{"cuisine":"Hamburgers"},{"cuisine":"American"}]})
df = json_normalize(cursor)
df.head()
```

```python
m = df.explode("grades")
q = m[["grades"]].apply(lambda r: r.grades ,result_type="expand", axis=1)

#display(m["grades"])
#display(q)
df2 = pd.concat([m,q],axis=1)
df2.head()
```

```python
df2 = df2.drop(columns=["grades"])
```

```python
df2.groupby(["restaurant_id","name"]).mean().reset_index().sort_values(by="score",ascending=False)
```
